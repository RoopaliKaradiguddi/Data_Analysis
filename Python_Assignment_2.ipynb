```python
df.drop(columns= ['index','Months'], inplace = True)
```

```python
df.isnull().sum()
```

```python
df[df.duplicated(['CUSTOMER','Style'], keep=False)]
```

```python
df.drop_duplicates(['Size','SKU'], inplace=True, ignore_index=True)
```

```python
df.isnull().sum()
```

```python
df['CUSTOMER'].fillna(0,inplace=True)
```

```python
df['Style'].fillna(0,inplace=True)
```

```python
df['SKU'].fillna(0,inplace=True)
```

```python
df['Size'].fillna(0,inplace=True)
```

```python
df['PCS'].fillna(0,inplace=True)
```

```python
df['RATE'].fillna(0,inplace=True)
```

```python
df['GROSS AMT'].fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

```python
cols = ["PCS", "RATE", "GROSS AMT"]

for col in cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.replace(",", "", regex=False)  # safety
    )
```

```python
df[cols] = df[cols].apply(pd.to_numeric, errors="coerce")
```

```python
df = df.dropna(subset=cols)
```

```python
df.dtypes
```

### DATA VISUALIZATION

### Sales by Size

```python
plt.figure(figsize=(23,8))

size_sales = df.groupby("Size")["GROSS AMT"].sum()

sns.barplot(
    x=size_sales.index,
    y=size_sales.values,
    palette="viridis"
)

plt.title("Sales by Size")
plt.xlabel("Size")
plt.ylabel("Total Gross Amount")
plt.show()
```

### Quantity Sold Over Time

```python
daily_qty = df.groupby("DATE")["PCS"].sum()

plt.figure(figsize=(20,8))
plt.plot(
    daily_qty.index,
    daily_qty.values,
    marker="o"
)
plt.xlabel("Date")
plt.ylabel("Quantity")
plt.title("Quantity Sold Over Time")
plt.show()
```

### Sales Trend Over Time

```python
daily_sales = df.groupby("DATE")["GROSS AMT"].sum()

plt.figure(figsize=(18,8))
plt.plot(daily_sales.index, daily_sales.values, marker="o", linewidth=2)

plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid(True)
plt.tight_layout()
plt.show()

```

### Distribution Analysis

```python
plt.figure(figsize=(10,5))
sns.histplot(
    df["GROSS AMT"],
    kde=True,
    color="pink"
)

plt.title("Sales Amount Distribution")
plt.show()
```

### Quantity Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(
    df["PCS"],
    kde=True,
    color="lightgreen"
)

plt.title("Quantity Distribution")
plt.show()
```

### Boxplot of Sales Amount

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    y=df["GROSS AMT"],
    color="red"
)

plt.title("Boxplot of Sales Amount")
plt.show()
```

### Top 10 SKU by Revenue

```python
top_sku = df.groupby("SKU")["GROSS AMT"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))
sns.barplot(
    x=top_sku.values,
    y=top_sku.index,
    palette="rocket"
)

plt.title("Top 10 SKUs by Revenue")
plt.xlabel("Total Sales")
plt.show()
```

### Quantity vs Gross Amount

```python
plt.figure(figsize=(10,5))
sns.scatterplot(
    x="PCS",
    y="GROSS AMT",
    data=df,
    color="Orange"
)

plt.title("Quantity vs Amount")
plt.show()
```
